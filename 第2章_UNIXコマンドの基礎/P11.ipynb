# タブをスペースに置換

タブ1文字につきスペース1文字に置換せよ．確認にはsedコマンド，trコマンド，もしくはexpandコマンドを用いよ．

```python
!cat hightemp.txt
```

```python
result = ''

with open('hightemp.txt') as f:
    result = f.read().replace('\t', ' ')
    
print(result)
```

## sed コマンド

僕使う OS 環境は macOS、 shell は fish shell なので、この sed コマンドは BSD 風なんです。どう使うのかさっぱりわからい、いろいろ調べたけど、

* https://qiita.com/mattintosh4/items/4e4d44016be15333af11
* https://qiita.com/atuyosi/items/12ceb5ca94231ce32d6e

結局諦めちゃった。

代わりに gnu-sed をインストールして、gsed を使いました。

インストール方法
`brew install gnu-sed`

```python
!gsed 's/\t/ /g' hightemp.txt
```

## tr コマンド

```python
!tr '\t' ' ' < hightemp.txt
```

## expand コマンド

```python
!expand -t 1 hightemp.txt
```
